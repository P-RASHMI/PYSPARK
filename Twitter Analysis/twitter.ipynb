```python
'''
@Author: Rashmi
@Date: 2021-11-24 18:20
@Last Modified by: Rashmi
@Last Modified time: 2021-11-24  23:57
@Title :Write a Python program for performing twitter analysis for real time
tweets and get the live graph for analysis using tweepy,textblob
'''
import findspark
findspark.init()
from LogHandler import logger
import requests
import csv
import os
import time
import tweepy
from textblob import TextBlob

from dotenv import load_dotenv
load_dotenv('.env')
```

```python
#to access all the authentication so as to access twitter API
consumer_key=os.getenv("consumer_key")
consumer_secret=os.getenv("consumer_secret")
access_token=os.getenv("access_token")
access_token_secret=os.getenv("access_token_secret")
```

```python
#connect to serverof twitter
auth = tweepy.OAuthHandler(consumer_key,consumer_secret)
```

```python
#connect to web sever of twitter
auth.set_access_token(access_token,access_token_secret)
```

```python
#connect to API storage server of twitter
api_connect = tweepy.API(auth)
```

```python
#to search particular word or topic on which we need analysis
#from twitter
tweet_word = api_connect.search_tweets("VIR DAS")
```

```python
pos='positive'
neg='negative'
pos_count=0
neg_count=0

list_headers = ["pos","neg","pos_count","neg_count"]
try:
    with open('tweetdata.csv','w') as csv_file:
        csv_obj = csv.DictWriter(csv_file,fieldnames=list_headers)
        csv_obj.writeheader()
    
    while True:
        #to append data continously line by line
        with open('tweetdata.csv','a') as csv_file:
            for tweet in tweet_word:
                # print(tweet.text)
                logger.info(tweet.text)
                #applying NLP(natural lang processing)
                analysis=TextBlob(tweet.text)
                # print(analysis.sentiment)
                logger.info(analysis.sentiment)
                #now checking polarity to differentiate positive,negative
                if analysis.sentiment.polarity > 0:
                    logger.info("positive")
                    print("positive")
                    pos_count = pos_count + 1
                elif analysis.sentiment.polarity < 0:
                    logger.info("negative")
                    print("negative")
                    neg_count = neg_count + 1
                else:
                    logger.info("neutral")
                csv_obj = csv.DictWriter(csv_file,fieldnames=list_headers)
                info_dict = {
                         "pos":pos,
                         "neg":neg,
                         "pos_count":pos_count,
                         "neg_count":neg_count,
                        }
                csv_obj.writerow(info_dict)
                time.sleep(1)
                print(pos_count,neg_count)
except Exception as e:
    logger.error(e)

```
